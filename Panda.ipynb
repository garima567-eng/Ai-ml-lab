```python
import pandas as pd
import matplotlib.pyplot as plt
Housing_price={
                    "Type":['Floor','Flat','Flat','Floor','House','Flat','House','Flat','Floor','Floor','House','Flat','Flat','Flat','House','Floor','Floor'],
                    "Size (in bhk)":[4,2,3,4,5,3,6,4,3,5,4,4,4,3,6,3,6],
                    "Area (in sq.ft)":[2500,650,750,2000,5000,700,5500,3700,670,980,760,1000,1200,870,4200,820,3800],
                    "Pricing (in crore)":[1.5,0.8,1,1.5,4.5,0.90,4,1,0.85,1.2,1,0.97,1,0.78,3,0.90,2]
}
df=pd.DataFrame(Housing_price)
p="Housing_price.csv"
df.to_csv(p,index=False)
```

```python
df=pd.read_csv('Housing_price.csv')
df["Size (in bhk)"]=(df["Size (in bhk)"]-df["Size (in bhk)"].min())/(df["Size (in bhk)"].max()-df["Size (in bhk)"].min())
df["Area (in sq.ft)"]=(df["Area (in sq.ft)"]-df["Area (in sq.ft)"].min())/(df["Area (in sq.ft)"].max()-df["Area (in sq.ft)"].min())
df["Pricing (in crore)"]=(df["Pricing (in crore)"]-df["Pricing (in crore)"].min())/(df["Pricing (in crore)"].max()-df["Pricing (in crore)"].min())
    
print(df.to_string())
```
