```python
class Problem:
    def __init__(self, initial, goal):
        self.initial = initial
        self.goal = goal

    def actions(self, state):
        pass

    def result(self, state, action):
        pass

    def goal_test(self, state):
        return state == self.goal
```

```python
class NumberProblem(Problem):
    def actions(self, state):
        return [1, 2]

    def result(self, state, action):
        return state + action
```

```python
from collections import deque

def bfs(problem):
    frontier = deque([problem.initial])
    explored = set()
    parent = {problem.initial: None}

    while frontier:
        state = frontier.popleft()

        if problem.goal_test(state):
            return parent

        explored.add(state)

        for action in problem.actions(state):
            child = problem.result(state, action)
            if child not in explored and child not in frontier:
                frontier.append(child)
                parent[child] = state

    return None
```

```python
problem = NumberProblem(initial=0, goal=5)
parents = bfs(problem)

# Reconstruct solution path
state = 5
path = []
while state is not None:
    path.append(state)
    state = parents[state]

print(path[::-1])
```

```python
graph={
    'S':{'A':1,'B':4},
    'A':{'C':2},
    'B':{'C':5},
    'C':{'G':3},
    'G':{}
    }



# DFS function to find path
def dfs(graph, current, goal, visited, path):
    visited.add(current)
    path.append(current)

    if current == goal:
        return True

    for neighbor in graph[current]:
        if neighbor not in visited:
            if dfs(graph, neighbor, goal, visited, path):
                return True

    path.pop()
    return False


# -------- Main Program --------


print("\nGraph:", graph)

# Start and goal from user
start = input("\nEnter start node: ")
goal = input("Enter goal node: ")

visited = set()
path = []

if dfs(graph, start, goal, visited, path):
    print("Path from", start, "to", goal, ":", " -> ".join(path))
else:
    print("No path found.")
```

```python
from collections import deque

graph = {}
# Function to add edges (bidirectional)
def add_edge(graph, u, v, weight):
    if u not in graph:
        graph[u] = {}
    if v not in graph:
        graph[v] = {}

    graph[u][v] = weight
    graph[v][u] = weight

    
def bfs(start, goal, graph):
    visited = set()
    queue = deque([[start]])

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)

            for neighbour in graph[node]:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)

    return None

n = int(input("Enter number of connections: "))

for i in range(n):
    u = input("Enter first node: ")
    v = input("Enter second node: ")
    w = int(input("Enter weight: "))
    add_edge(graph, u, v, w)

print("\nGraph:", graph)

start = input("Enter start node: ")
goal = input("Enter goal node: ")

result = bfs(start, goal, graph)

if result:
    print("BFS Path:", " -> ".join(result))
else:
    print("No path found")
```
